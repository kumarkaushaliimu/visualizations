```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('GlobalSuperstoreData.csv')
```

```python
df.tail()
```

```python
# convert to date
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# verify datatype
print(type(df['Order Date'].iloc[0]))
```

```python
df =df.sort_values(by='Order Date')
```

```python
df.head()
```

**Time** **realted** **plots**

```python
#Grouping sales and profit as per dates
```

```python
df_sales = df.groupby(['Order Date'],as_index=False).sum()
df_sales.head()
```

```python
fig = px.line(df_sales, x='Order Date', y="Sales",markers =True,title='Sales time series')
fig.show()
```

```python
#Profit time series
fig = px.line(df_sales, x="Order Date", y="Profit",markers =True,title='Profit time series')
fig.show()
```

```python
#Segment wise Sales data
df_segment = df.groupby(['Order Date','Segment'],as_index=False).sum()
fig = px.line(df_segment, x="Order Date", y="Sales",markers =True,title='Sales time series-Segment wise',color = 'Segment')
fig.show()
```

```python
#Segment wise Profit data
fig = px.line(df_segment, x="Order Date", y="Profit",markers =True,title='Profit time series-Segment wise',color = 'Segment')
fig.show()
```

```python
#Market wise Sales data
df_market = df.groupby(['Order Date','Market'],as_index=False).sum()
fig = px.line(df_market, x="Order Date", y="Sales",markers =True,title='Sales time series-Market wise',color = 'Market')
fig.show()
```

```python
#Market wise Profit data

fig = px.line(df_market, x="Order Date", y="Profit",markers =True,title='Profit time series-Market wise',color = 'Market')
fig.show()
```
